```python
heart_disease = pd.read_csv('data/heart-disease.csv')
```

```python
heart_disease.head(10)
```

```python
plt.hist(heart_disease['age'], bins=50);
```

```python
heart_disease.plot.hist();
```

```python
heart_disease.plot.hist(subplots=True, figsize=(10, 30));
```

```python
hearts50 = heart_disease[heart_disease["age"] > 50]
```

```python
plt.style.use('seaborn-whitegrid')
fig, ax = plt.subplots(figsize=(10, 6))

# colors = np.array(["black", "green"])

JessicaDiaz = ax.scatter(x=hearts50["age"], 
            y=hearts50["chol"],
            c=hearts50["target"],
            cmap="winter")

mean_line = ax.axhline(hearts50["chol"].mean(), 
           linestyle='--',
           color="red")

ax.set(title="Hearts 50 yrs",
      xlabel="age",
      ylabel="cholesterol")
handles, fruit = JessicaDiaz.legend_elements()
legend1 = ax.legend([handles[0], handles[1], mean_line], 
                    [fruit[0], fruit[1], "mean"], 
                    title="Target");
```
